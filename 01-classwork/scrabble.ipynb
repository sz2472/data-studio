```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib
```

```python
points_df  = pd.read_csv("scrabble-point-spread.csv")
tournament_df = pd.read_csv("scrabble-tournament.csv")
df = tournament_df.merge(points_df, left_on='Player ID', right_on='player_id').fillna(0)
df.head()
```

```python
df.columns
```

```python
df_division_1 = df[df["division"]=="Division 1"]
df_division_2 = df[df["division"]=="Division 2"]
df_division_3 = df[df["division"]=="Division 3"]
df_division_4 = df[df["division"]=="Division 4"]
```

```python
plt.style.use("fivethirtyeight")
```

```python
ax = df_division_1.plot(kind="scatter", y='points_against', x='points_for', label='Division 1', figsize=(10,10))
df_division_2.plot(kind="scatter", y='points_against', x='points_for', ax=ax, label='Division 2', color='red')
df_division_3.plot(kind="scatter", y='points_against', x='points_for', ax=ax, label='Division 3', color='yellow')
df_division_4.plot(kind="scatter", y='points_against', x='points_for', ax=ax, label='Division 4', color='green')

ax.set_xlim([300, 460])
ax.set_ylim([300, 460])
ax.legend()

ax.set_title("Nigle Richard's Performance At Nationals")
ax.set_ylabel("Points Against")
ax.set_xlabel('Points For')

plt.savefig("Nigle-Richard's-Performance-At-Nationals.pdf", transparent = True)
```
