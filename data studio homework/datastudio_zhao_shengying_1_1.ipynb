```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import matplotlib
```

```python
point_df  = pd.read_csv("scrabble-point-spread.csv")
tournament_df = pd.read_csv("scrabble-tournament.csv")
```

```python
point_df.head()
```

```python
tournament_df.head()
```

```python
df = tournament_df.merge(point_df, left_on='Player ID', right_on='player_id').fillna(0)
```

```python
df.head()
```

```python
df_1 = df[df["division"]=="Division 1"]
df_2 = df[df["division"]=="Division 2"]
df_3 = df[df["division"]=="Division 3"]
df_4 = df[df["division"]=="Division 4"]
```

```python
fig, ax = plt.subplots(figsize=(10,5))
for category, group in df.groupby('division'):
    ax.plot(group['points_for'],group['points_against'], marker='o', linestyle='', label=category, markeredgewidth=0)
    ax.set_xlim([300,460])
    ax.set_ylim([300,460])
    ax.legend()
plt.savefig("scrabble.pdf",transparent=True)
```
